```python
import pandas as pd

# Vars
gpa_dict = dict()
course = 'null'
courses = []
course_credit = 0
course_credits_list = []
course_grade = 0.0
course_grade_list = []
raw_points_list = []
prior_creds = 0
prior_gpa = 0.0
```

```python
# Run this block if this isn't your first semester at the University, otherwise, skip if it is or it's your first xfer semester

prior_creds = int(input('Input the number of previously completed credits you have: '))
prior_gpa = float(input('Input your current cumulative GPA: '))
```

```python
# Run this block to enable the raw value fetching part of the calculator

def get_value(grade):
    """Converts letter grade to GPA value (float) to be used in calculating final grade point average.
    Values from https://www.testudo.umd.edu/gpa/index.html"""
    value = 0.0
    grade = grade.lower()
    if grade == 'a' or grade == 'a+':
        value = 4.0
    elif grade == 'a-':
        value = 3.7
    elif grade == 'b+':
        value = 3.3
    elif grade == 'b':
        value = 3.0
    elif grade == 'b-':
        value = 2.7
    elif grade == 'c+':
        value = 2.3
    elif grade == 'c':
        value = 2.0
    elif grade == 'c-':
        value = 1.7
    elif grade == 'd+':
        value = 1.3
    elif grade == 'd':
        value = 1.0
    elif grade == 'd-':
        value = 0.7
    else:
        value = 0.0
    return value
```

```python
# Manual Input Mode
print('Follow line by line instructions, and then input DONE to finish.')
while True:
    course = (input('Enter a course designator (ex. MATH 140): '))
    if course == 'done':
        break
    courses.append(course)
    course_credits = int((input('Enter course credit value (integer): ')))
    course_credits_list.append(course_credits)
    course_grade = (input('Enter your letter grade achieved in this class: '))
    course_grade_list.append(course_grade)
    raw_points_list.append(float(course_credits) * float(get_value(course_grade)))

# Dataframe visualization for Manual Input Mode
df = pd.DataFrame()
df['Courses'] = courses
df['Credits'] = course_credits_list
df['Grade'] = course_grade_list
df['Raw_Points'] = raw_points_list

df
```

```python
# If you want to save your manually inputted progress for later adjustments, run this block to output the Dataframe to a .csv
# Accounts for the column headers and leaves out the iterative row index

df.to_csv('gpa_reference_list.csv', header = True, index = False)
```

```python
# CSV Input Mode/Visualization
filename = input('Input your complete filename and extension (ex. filename.csv, .csv only): ')
df = pd.read_csv('{}'.format(filename))
for key,value in df['Grade'].iteritems():
    raw_points_list.append(get_value(value))
df['Raw_Score'] = raw_points_list
df['Raw_Points'] = df['Credits'] * df['Raw_Score']
df
```

```python
# Final GPA Calculation for either Input Method
new_credit_total = sum(df['Credits'])

# final_credits calculates the final amount of credits that the total amount of raw points will be divided by to calculate GPA
final_credits = prior_creds + int(new_credit_total)

# Prior weighted GPA points to be factored in
weighted_prior_gpa = prior_creds * prior_gpa

# Final simple equation to calculate the final GPA
final_gpa = (weighted_prior_gpa + sum(df['Raw_Points']))/final_credits
print(final_gpa)
```
